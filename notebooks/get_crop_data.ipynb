```python
# In order for this shortcut filepath to work I had to allow full disk access to the terminal (since that's how I was launching jupyter notebook)
crop_data = pd.read_csv(file_path)
```

```python
# Filter down to just PT
pt_data = crop_data[crop_data['country'] == "Portugal"]
```

```python
columns_to_drop = [col for col in pt_data.columns if col.endswith("_attainable_yield") or col.endswith("_yield_gap")]
pt_filtered = pt_data.drop(columns=columns_to_drop)
```

```python
columns_any_non_na = pt_filtered.columns[pt_filtered.notna().any()].tolist()
```

```python
df_cleaned = pt_filtered.dropna(axis=1, how='all')
```

```python
# Show the max number of columns possible
pd.set_option('display.max_columns', None)
# Yields are measured in tonnes per hectare
df_cleaned.head(50)
```

```python
# Write to .csv
today = datetime.now().strftime("%Y%m%d")
file_name = "/Users/eleshuk/Library/CloudStorage/GoogleDrive-eleshuk@gmail.com/.shortcut-targets-by-id/1XNkQR60z1T7WELqvqkvZchRrVWQpCzvs/data_science_project/Project Template/data/processed/"+today+"_crop_yield_cleaned.csv"
df_cleaned.to_csv(file_name, sep=',', encoding='utf-8')
```

```python
# Convert from wide to long format
melted_df = df_cleaned.melt(id_vars=['country', 'year'], var_name='crop', value_name='yield')
```

```python
melted_df = melted_df.dropna()
# melted_df['crop'].unique()
```

```python
melted_df['crop'] = melted_df['crop'].str.replace('_yield', '')
```

```python
# Write to .csv
today = datetime.now().strftime("%Y%m%d")
file_name = "/Users/eleshuk/Library/CloudStorage/GoogleDrive-eleshuk@gmail.com/.shortcut-targets-by-id/1XNkQR60z1T7WELqvqkvZchRrVWQpCzvs/data_science_project/Project Template/data/processed/"+today+"_crop_data_long.csv"
melted_df.to_csv(file_name, sep=',', encoding='utf-8')
```

```python
top_20_crops = melted_df.sort_values(['year', 'yield'], ascending=[True, False]) \
                        .groupby('year') \
                        .head(20) \
                        .sort_values('year')
```

```python
# Write to .csv
today = datetime.now().strftime("%Y%m%d")
file_name = "/Users/eleshuk/Library/CloudStorage/GoogleDrive-eleshuk@gmail.com/.shortcut-targets-by-id/1XNkQR60z1T7WELqvqkvZchRrVWQpCzvs/data_science_project/Project Template/data/processed/"+today+"_top_20_crops_by_year.csv"
top_20_crops.to_csv(file_name, sep=',', encoding='utf-8')
```

```python
# Make a list of unique crops
sorted_unique_values = sorted(melted_df['crop'].unique())
unique_crops_df = pd.DataFrame(sorted_unique_values, columns=['crop'])
```
